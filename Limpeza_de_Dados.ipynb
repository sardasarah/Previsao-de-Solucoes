```python
import pandas as pd
df = pd.read_csv("/content/dataset-limpo2")
```

```python
df
```

```python
df['defeito_constatado_descricao'] = df['defeito_constatado_descricao'].str.title()
```

```python
df['familia_descricao'] = df['familia_descricao'].str.title()
```

```python
df
```

```python
df['solucao_descricao'].unique()
```

```python
correcoes = {
    'Troca De Rele': 'Troca De Relé',

    'Troca De Fonte 12/24V': 'Troca Da Fonte 12/24 V',
    'Reparo Na Fonte 12/24V': 'Reparo Da Fonte 12/24 V',

    'Troca/Reparo Unidade De Refrig': 'Troca/Reparo Unidade De Refrigeração',
    'Troca De Unidade De Refrig': 'Troca/Reparo Unidade De Refrigeração'
}

df['solucao_descricao'] = df['solucao_descricao'].replace(correcoes)
```

```python
df
```

```python
df['solucao_descricao'].unique()
```

```python
df["categoria_defeito"] = df["categoria_defeito"].replace(
    "eletrônico",
    "Eletrônico"
)

df = df[
    df["categoria_defeito"] !=
    "À medida que avalanche de produtos químicos avançava, a maioria já estava absorvida."
]

df["categoria_defeito"] = df["categoria_defeito"].replace(
    "Moedeiro /",
    "Moedeiro"
)
```

```python
"À medida que avalanche de produtos químicos avançava, a maioria já estava absorvida." in df["categoria_defeito"].values
```

```python
df["familia_descricao"] = df["familia_descricao"].replace(
    "Maquina De Cafe",
    "Máquina de Café"
)
```

```python
df.to_csv('dataset-limpo-projeto.csv', index=False)
```
